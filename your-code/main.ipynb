#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
steps_month = fitbit.groupby("Months_encoded")["Steps"].median()
```

```python
steps_day = fitbit.groupby("Work_or_Weekend")["Steps"].median()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
week_work = ['Work', 'Weekend']
```

```python
#plot
fig_5, ax_5 = plt.subplots(1, 2)

# MEDIAN STEPS BY MONTH_ENCODED
plt.sca(ax_5[0])
ax_5[0].plot(steps_month.index, steps_month, label="by month", c = "darkblue") 
plt.xticks(range(1,13),months)

# MEDIAN STEPS BY WORK_OR_WEEKEND
plt.sca(ax_5[1])
ax_5[1].bar(x = steps_day.index, height = steps_day, label="workday or weekend", color = ['teal', 'aqua'])
plt.xticks(range(2),week_work)

# Title
fig_5.suptitle("Fitbit data", fontsize= 15) 

#legend
ax_5[0].legend()
ax_5[1].legend()

#Size
fig_5.set_figheight(5)
fig_5.set_figwidth(12)
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fig_6, ax_6 = plt.subplots(1, 3, figsize = (15,10))

columns = ["Minutes Lightly Active", "Minutes Very Active", "Minutes Sedentary"]

for i, c in enumerate(columns):
    ax_6[i].scatter(x = fitbit[c], y = fitbit["Steps"], label = c, color = "orange")
    ax_6[i].legend()
    ax_6[i].set_title(c)
    
    
```
